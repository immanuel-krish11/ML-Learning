```python
# ______workflow_______

# 1.import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

# 2. prepare data
x = [1,2,3,4]
y = [2,4,6,7]
# 3. setup plot
fig, ax = plt.subplots(figsize = (10,10))  # figsize tells height width 

# 4. plot data
ax.plot(x,y)

# 5. customise plot
ax.set(title = "simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 6. save and show

fig.savefig("sample-plot.png")
```

```python
# histograms 
x = np.random.randn(1000)

fig, ax = plt.subplots()

ax.hist(x)
plt.hist(x)
plt.show()
```
