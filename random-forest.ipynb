```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(['No', 'Yes'])
```

```python
df['smoker'].replace({'Yes': 1, 'No':0}, inplace=True)
```

```python
df
```

```python
df_encode = pd.get_dummies(df, drop_first=True)
df_encode
```

```python
df_encode[['sex_Female', 'smoker_0', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Dinner']] = df_encode[['sex_Female', 'smoker_0', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Dinner']].astype(int)
df_encode
```

```python
x = df_encode.drop('tip', axis = 1)
y = df_encode['tip']
```

```python
x
y
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
x_train_transformed = ohe.fit_transform(x_train[['sex','day', 'time']])
```

```python
x_train_transformed.shape
```

```python
x_test_transformed = ohe.fit_transform(x_test[['sex','day', 'time']])
x_test_transformed
```

```python
from sklearn.ensemble import RandomForestRegressor
model_regressor = RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'criterion': ['squared_error', 'absolute_error'],
    'max_depth': [1,2,3,4,5,6,7],
}

grid_search = GridSearchCV(estimator=model_regressor, param_grid=params, cv = 5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
y_pred = grid_search.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
score = mean_squared_error(y_pred, y_test)
score
```
